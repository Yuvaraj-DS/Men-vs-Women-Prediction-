# Convert Images to Arrays

```python
import os
import numpy as np
import cv2
```

```python
DIRECTORY = r'C:\Users\yuvi\Documents\ML Data Analysis\FACE'
CATEGORIES = ['men', 'women']
```

```python
DIRECTORY
```

```python
os.path.join(DIRECTORY, CATEGORIES[0])
```

```python
import time
start_time = time.time()

data = []

IMG_SIZE = 75


for category in CATEGORIES:
    path = os.path.join(DIRECTORY, category)
    for image in os.listdir(path):
        try:
            gender = CATEGORIES.index(category)
            img_path = os.path.join(path, image)
            arr = cv2.imread(img_path)
            arr = cv2.resize(arr, (IMG_SIZE, IMG_SIZE))
            data.append([arr, gender])
        except Exception as e:
            print(e)
            pass

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
len(data)
```

```python
import random

random.shuffle(data)
```

```python
X = []
y = []

for features, label in data:
    X.append(features)
    y.append(label)
```

```python
y
```

```python
X = np.array(X)
```

```python
X = X/255
```

```python
X
```

```python
X = X.reshape(-1, IMG_SIZE, IMG_SIZE, 3)
```

```python
X.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
```

```python
from sklearn.model_selection import train_test_split
X_train, y_train, X_test, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
model = Sequential()
#model.add(Conv2D(filters = 128, kernel_size = (3, 3)))
model.add(Conv2D(64, kernel_size=1, input_shape=(75, 75, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

#model.add(Dense(100, activation = 'relu'))
#model.add(Dense(100, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=3, validation_data=(X_test, y_test))
```
